## load data and preprocessing

```python
## load and conversion

data = pd.read_csv('data/train.tsv', sep='\t', na_values=['?', 'null'])
#data = pd.read_csv('data/test.tsv', sep='\t', na_values=['?', 'null'])
data['boilerplate'] = data['boilerplate'].apply(json.loads)
data['title'] = data['boilerplate'].apply(lambda x: x.get('title', np.nan))
data['body'] = data['boilerplate'].apply(lambda x: x.get('body', np.nan))
data['domain'] = data['url'].apply(lambda x: '.'.join(urlsplit(x).netloc.split('.')[0:]))
data['embed_ratio'].loc[data['embed_ratio'] < 0] = np.nan 

data.loc[pd.isnull(data.alchemy_category), 'alchemy_category'] = 'unknown'
```

```python
domain_counts = sorted([(domain, df.shape[0]) for domain, df in data.groupby('domain')],
                       key = lambda x: x[1], reverse = True)
domain_counts = dict(domain_counts)

data['rough_domain'] = data['domain'].apply(lambda x: x if domain_counts[x] > 30 else 'others')
```

## data matrix

```python
selected_discrete_features = ['alchemy_category', u'rough_domain']
selected_continuous_features = ['avglinksize',
            u'commonlinkratio_1', u'commonlinkratio_2', u'commonlinkratio_3', 
            u'commonlinkratio_4', u'compression_ratio',  
            u'frameTagRatio',  u'html_ratio', u'image_ratio', 
            u'linkwordscore',  
            u'non_markup_alphanum_characters', 
            u'numberOfLinks', u'numwords_in_url', 
            u'parametrizedLinkRatio', u'spelling_errors_ratio']
selected_text_features =['title', 'body']
```

```python
## missing values
for f in selected_discrete_features+selected_continuous_features+selected_text_features:
    print f, sum(pd.isnull(data[f]))
```

```python
## discrete encoders
alchemy_category_labeler = LabelBinarizer().fit(data.alchemy_category.tolist())
encoded_alchemy_category = alchemy_category_labeler.transform(data.alchemy_category.tolist())
print encoded_alchemy_category.shape

rough_domain_labeler = LabelBinarizer().fit(data.rough_domain.tolist())
encoded_rough_domain = alchemy_category_labeler.transform(data.rough_domain.tolist())
print encoded_rough_domain.shape
```

```python
X = np.hstack([data.loc[:, selected_continuous_features], 
               encoded_alchemy_category, 
               encoded_rough_domain])
feature_names = (selected_continuous_features 
                 + alchemy_category_labeler.classes_.tolist() 
                 + rough_domain_labeler.classes_.tolist())
y = np.asarray(data.loc[:, 'label'])
print X.shape, y.shape
```

## baseline modelling on non-text features

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import KFold
from sklearn.grid_search import GridSearchCV, RandomizedSearchCV
```

```python
estimator = ExtraTreesClassifier(random_state=0)
param_grid = {'n_estimators': [50, 100, 200, 300, 500], 
              'max_features': ['auto', 2, 5, 10, 15, 20, 30]}
cv = KFold(y.shape[0], n_folds = 3, shuffle=True, random_state = 0)
gs = GridSearchCV(estimator, param_grid, cv = cv, n_jobs = -1)
%time gs.fit(X, y)
print gs.best_params_
print gs.best_score_
```

```python
## feature importances
trees = ExtraTreesClassifier(n_estimators=200, max_features=10, oob_score = True, 
                             bootstrap=True, random_state=0)
trees.fit(X, y)
print trees.oob_score_ ## at least higher than 0.51 
```

```python
t = trees.estimators_[0].tree_
t.compute_feature_importances()
```

```python
feature_importances = utils.plot_feature_importances(trees, feature_names)
feature_importances
```

```python
useful_feature_names = [f for f,s in feature_importances if s > 0]
useful_features_index = [feature_names.index(f) for f in useful_feature_names]
useful_X = X[:, useful_features_index]
print useful_feature_names
print useful_X.shape
```

```python
## use selected features ONLY

estimator = ExtraTreesClassifier(random_state = 0)
param_grid = {'n_estimators': [50, 100, 200, 300, 500], 
              'max_features': ['auto', 2, 5, 10, 15, 20, 25]}
cv = KFold(y.shape[0], n_folds = 3, shuffle=True, random_state = 0)
gs = GridSearchCV(estimator, param_grid, cv = cv, n_jobs = -1)
%time gs.fit(useful_X, y)
print gs.best_params_
print gs.best_score_
```

```python
## how complicated is the model copmared to a LINEAR
## it must be NORMALIED BEFORE USING SVM

from sklearn.preprocessing import normalize, scale
from sklearn.svm import LinearSVC, SVC

#normalized_X = normalize(useful_X) # NOT really good for one hot encoding
normalized_X = scale(useful_X)

estimator = LinearSVC(random_state = 0)
param_grid = {'C': [1, 10, 100, 500, 1000]}
cv = KFold(y.shape[0], n_folds = 3, shuffle=True, random_state = 0)
gs = GridSearchCV(estimator, param_grid, cv = cv, n_jobs = -1)
%time gs.fit(normalized_X, y)
print gs.best_params_
print gs.best_score_
```

```python
estimator = SVC(random_state = 0)
param_grid = {'C': [1, 10, 100, 500, 1000, 10000],
              'gamma': [0.0001, 0.001, 0.01, 0.1]}
cv = KFold(y.shape[0], n_folds = 3, shuffle=True, random_state = 0)
gs = GridSearchCV(estimator, param_grid, cv = cv, n_jobs = -1)
%time gs.fit(normalized_X, y)
print gs.best_params_
print gs.best_score_
```
